```python
whr_merged = whr_2018.merge(whr_2019, left_on = "Country or region", right_on = "Country or region", suffixes = ["_2018", "_2019"])
# combines the two datasets into a single dataset based on the specified key column                                
```

```python
scores = whr_merged[["Country or region", "Score_2018", "Score_2019"]]
# ctreate a new variable from columns of whr_merged
```

```python
scores
# show the result
```

```python
scores.nlargest(20, "Score_2018")
# show the largest values in the "Score_2018" column
```

```python
plt.barh(scores.nlargest(20, "Score_2018")["Country or region"], scores.nlargest(20, "Score_2018")["Score_2018"])
plt.gca().invert_yaxis()
plt.xlim(0, 8)
plt.show()
# create a horizontal bar for the top 20 countries with the largest "Score_2018" values from the scores
# when use bar we submit 2 variables
```

```python
plt.figure(figsize = (8, 10))

plt.barh(scores.nsmallest(20, "Score_2018")["Country or region"], scores.nsmallest(20, "Score_2018")["Score_2018"])
plt.xlim(0, 8)
plt.show()
# create a new figure object with a specified figure size
```

```python
whr_merged = whr_merged.rename(columns = {"Happiness.Score_2017": "Score_2017"})
# rename of one column
```

```python
whr_merged
# show the results
```

```python
scores = whr_merged[["Country or region", "Score_2017", "Score_2018", "Score_2019"]]
# ctreate a new dataset with 3 columns
```

```python
bg_scores = scores[scores["Country or region"] == "Bulgaria"].T.drop("Country or region").values.ravel()
ro_scores = scores[scores["Country or region"] == "Romania"].T.drop("Country or region").values.ravel()
# extract scores for Bulgaria and Romania from the scores dataset, transpose the data,
# drop the "Country or region" column, and then flatten the values into a 1-dimensional array with ravel()
```

```python
years = list(range(2017, 2020))
# create a list of years from 2017 to 2019 (inclusive)
```

```python
plt.scatter(years, bg_scores, linewidth = 2, linestyle = "dashed", label ="BG")
plt.plot(years, bg_scores, marker = "s")

plt.scatter(years, ro_scores, linewidth = 2, linestyle = "dashed", label = "RO")
plt.plot(years, ro_scores, marker = "s")
plt.ylim(0, 6.5)

# plt.ylim(0, np.max(ro_scores) +0.25)
#namirane na max na ro_scores

plt.title("Bulgaria: Happiness score")
plt.xlabel('Year')
plt.ylabel("Score")
plt.xticks(years)
plt.legend(loc = "lower right")
plt.show()
# 
```
